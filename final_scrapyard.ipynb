```python
from flask import Flask, render_template, request, jsonify
from pyngrok import ngrok
from textblob import TextBlob
import random

app = Flask(__name__)

# Start ngrok
ngrok.set_auth_token("2ugaR0Y3InBsUr5dLQlL8pMwp8d_2MAVeMffzYt7iXFXv4xzn")
ngrok_tunnel = ngrok.connect(5000)
print('Public URL:', ngrok_tunnel.public_url)

# Snake facts and motivational quotes
snake_facts = [
    "Did you know some snakes can survive without food for months?",
    "Cobras can spit venom up to 8 feet. That’s further than your gym coach’s throw!",
    "Snakes have over 400 vertebrae, which makes them ultra-flexible.",
    "A snake’s jaw can unhinge to swallow prey whole. Imagine doing that with a large pizza."
]

motivational_quotes = [
    "Shedding your skin, like a snake, is a part of growth. Embrace it!",
    "Not all constriction is bad — sometimes it’s just a hug from life.",
    "Even the tiniest garden snake has the spirit of a python. Keep going!",
    "Snakes can’t dwell on their past sheds, and neither should you."
]

#roast or backhanded compliment
def roast_or_backhanded():
  choice = random.choice(["roast", "backhanded"])
  roast = ["You bring everyone so much joy... when you leave the room.", "You have something on your chin… no, the third one.",
"You’re like a cloud. When you disappear, it’s a beautiful day.",
"You have something on your face. Oh, never mind—it’s just your face.",
"You’re proof that even evolution takes the occasional coffee break."
"You bring a lot to the table. Unfortunately, it’s mostly baggage.",
"You have the perfect face for radio.",
"You’re like a software update—nobody actually wants you, they just tolerate you out of necessity.",
"I’d agree with you, but then we’d both be wrong.",
"You have such a magnetic personality. Too bad it only repels.",
"I’d explain it to you, but I left my crayons at home.",
"You have something that not everyone has: a pulse. Let’s not push it.",
"You remind me of a cloud. Fluffy, shapeless, and easy to ignore.",
"You’re like a broken traffic light—no one knows when to stop around you.",
"You have such a creative mind. It’s a shame you only use it for excuses.",
"You bring a lot of energy to the table. Pity it’s all misplaced.",
"You have the unique ability to say nothing and still be wrong.",
"You’re like a phone with 1% battery. Unreliable and constantly draining.",
"I’d say you’re sharp, but even a broken pencil has more point.",
"You have such a warm personality. Like a car with no A/C in July."]
  backhanded = ["You have such a gift for making simple things... complicated.",
"I love how you’re always the main character, no matter whose story it is." ,
"You’re so good at giving advice you clearly don’t follow.",
"You bring such a unique energy to the room—like when someone forgets to turn off their alarm.",
"You’re so down-to-earth. It’s almost like gravity’s working extra hard on you.",
"I admire how you never let talent get in the way of enthusiasm.",
"You have this amazing ability to always make things about you—it’s honestly impressive.",
"You’re so effortlessly unpredictable. Like a Wi-Fi signal.",
"I wish I had your confidence. Especially when you’re wrong.",
"You always know how to make an entrance. And an exit. Sometimes both at once.",
"It’s incredible how you can turn any conversation into a TED Talk. Even when no one asked.",
"You have such a distinctive laugh. It really... lingers.",
"You’re like a human autocorrect—always offering suggestions no one asked for.",
"I love how you’re so committed to your opinions. Even when they make no sense.",
"You have such a rare talent for making things harder than they need to be.",
"You’re so refreshingly direct. Like a frying pan to the face.",
"You have this amazing ability to stay calm. Especially when you probably shouldn’t be.",
"I admire how you never let the facts ruin a perfectly good argument.",
"You’re such a trendsetter. I’ve never seen anyone misinterpret things quite like you.",
"I love how you can light up a room. Even if it’s with chaos."]
  if choice == "roast":
    return(random.choice(roast))
  else:
    return(random.choice(backhanded))


def productivity_assigner():
    return f"You have been {random.randint(1, 100)}% productive today. Your effort has been acknowledged by the Council of Bees."

# life advice
def life_advice():
  import random
  templates = [
    "{} once said, '{}'. Take that with a grain of cosmic dust.",
    "If you ever find yourself {}, remember to {}.",
    "Some say that {} is the key to {}, but don't forget to {}.",
    "A true philosopher knows that {} is simply {} in disguise."]

  words1 = ["A sentient cactus", "The spirit of a confused jellybean", "Your refrigerator", "A very wise pigeon"]
  words2 = ["'Life is just a series of questionable snack choices.'", "'Always butter your toast on both sides.'", "'Laughter is like a hiccup for the soul.'"]
  words3 = ["hugging a lamppost", "arguing with a cloud", "dancing with the shadows"]
  words4 = ["embrace the chaos", "question your socks", "sing to your furniture"]
  words5 = ["tickling the void", "becoming one with the spaghetti", "shouting at the moon"]
  words6 = ["the meaning of Tuesdays", "the forbidden art of reverse cooking", "a very specific shade of regret"]

  def generate_life_advice():
      template = random.choice(templates)
      if "{} once said" in template:
          return template.format(random.choice(words1), random.choice(words2))
      elif "find yourself" in template:
          return template.format(random.choice(words3), random.choice(words4))
      elif "the key to" in template:
          return template.format(random.choice(words5), random.choice(words6), random.choice(words4))
      else:
          return template.format(random.choice(words1), random.choice(words6))

  return(generate_life_advice())


def meal_name():
  breakfast = ['waffle', 'bacon', 'french toast', 'cereal', 'hash browns', 'pancakes', 'bagel']
  lunch = ['salads', 'bacon', "mac 'n cheese", 'chicken broth', 'cheeseburger', 'fettuccine pasta', 'lasagna', 'pizza']
  afternoon_tea = ['cucumber sandwiches', 'scones', 'macarons', 'cupcakes', 'eclairs']
  dinner = ['pizza', 'beans and rice', 'stir fry', 'fajitas', 'quiche']
  dessert = ['chocolate cake', 'fudge brownies', 'banana ice cream', 'strawberry cheesecake', 'chocolate chip cookies']
  group1 = random.sample([breakfast, lunch, afternoon_tea, dinner, dessert], 2)
  return("I predict that you are "+random.choice(['sad', 'happy', 'frustrated', 'disgusted']) + " so you ought to eat "+random.choice(group1[0])+ " and " + random.choice(group1[1])+".")


# Sentiment analysis for snake therapist
def analyze_sentiment(text):
    analysis = TextBlob(text)
    sentiment_score = analysis.sentiment.polarity
    if sentiment_score > 0:
        return "positive"
    elif sentiment_score < 0:
        return "negative"
    else:
        return "neutral"

# Dice route
@app.route('/roll', methods=['GET'])
def roll_dice():
    dice_roll = random.randint(1, 6)

    if dice_roll == 1:
        result = life_advice()
    elif dice_roll == 2:
        result = meal_name()
    elif dice_roll == 3:
        result = roast_or_backhanded()
    elif dice_roll == 4:
        result = productivity_assigner()
    elif dice_roll == 5:
        result = "You got a special roll! Talk to the snake therapist 🐍!"
    else:
        result = "A loud honk echoes in your mind. That's it. Just honk."

    return jsonify({"roll": dice_roll,"result": result})

# Snake therapist route
@app.route('/snake', methods=['POST'])
def snake_therapist_api():
    user_input = request.form.get('user_input')

    if not user_input:
        return jsonify({"response": "Hiss... tell me something."})

    sentiment = analyze_sentiment(user_input)

    if sentiment == "positive":
        response = "Good to see that you are happy. Have a random snake fact: " + random.choice(snake_facts)
    elif sentiment == "negative":
        response = random.choice(motivational_quotes)
    else:
        response = "I sense neutrality. Did you know? " + random.choice(snake_facts)

    return jsonify({"response": response})

@app.route('/')
def index():
    return render_template('index.html')

if __name__ == '__main__':
    app.run()
```
